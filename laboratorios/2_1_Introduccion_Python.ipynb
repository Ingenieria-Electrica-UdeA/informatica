**Funcin `print()`**

Similarmente se usa la funcin `print()`

```python
print('La edad de', nombre, 'es', edad, 'aos')
```

Es posible especificar el separador entre partes de la cadena de caracteres

```python
print('La edad de',nombre,'es',edad,'aos',sep=' - ')
```

```python
print('La edad de',nombre,'es',edad,'aos',sep='\t\t')
```

```python
print('La edad de',nombre,'es',edad,'aos',sep='\n')
```

Tambin es posible construir el texto agregando al texto esttico  los lugares dnde deseamos incormporar informacin proveniente de variables usando el formato `%.Xf` dnde % indica dnde ira localizada una variable y Xf su formato. Por ejemplo `.4f` significa que el nmero contendr cuatro cifras significativas

```python
pi
```

```python
print('pi = ', pi)
```

```python
print('pi = %.4f' % pi)
```

```python
pi
```

```python
raiz = np.sqrt(3)
raiz
```

```python
print('pi = %.4f, y la raz de 3 es: %.3f' % (pi, raiz))
```

```python
print('pi = %.4f, y la raz de 3 es: %d' % (pi, raiz))
```

```python
print('La edad de %s es %d aos' % (nombre, edad))
```

Un resultado similar se logra con los denominados placeholders `{}`

```python
nota = 3.45
estudiante = "Pablo"

print("La nota de {} es {:.2f}".format(estudiante, nota))
```
