### <font color='46B8A9'> **Salida** </font>

Se pueden usar cadenas de caracteres con un prefijo: `f'mensaje + variables'`
<br>


```python
nombre = 'Facundo'
edad = 53
```

```python
f'La edad de {nombre} es {edad} años'
```

**Función `print()`**

Similarmente se usa la función `print()`

```python
print('La edad de', nombre, 'es', edad, 'años')
```

Es posible especificar el separador entre partes de la cadena de caracteres

```python
print('La edad de',nombre,'es',edad,'años',sep=' - ')
```

```python
print('La edad de',nombre,'es',edad,'años',sep='\t\t')
```

```python
print('La edad de',nombre,'es',edad,'años',sep='\n')
```

También es posible construir el texto agregando al texto estático  los lugares dónde deseamos incormporar información proveniente de variables usando el formato `%.Xf` dónde % indica dónde ira localizada una variable y Xf su formato. Por ejemplo `.4f` significa que el número contendrá cuatro cifras significativas

```python
pi
```

```python
print('pi = ', pi)
```

```python
print('pi = %.4f' % pi)
```

```python
pi
```

```python
raiz = np.sqrt(3)
raiz
```

```python
print('pi = %.4f, y la raíz de 3 es: %.3f' % (pi, raiz))
```

```python
print('pi = %.4f, y la raíz de 3 es: %d' % (pi, raiz))
```

```python
print('La edad de %s es %d años' % (nombre, edad))
```

Un resultado similar se logra con los denominados placeholders `{}`

```python
nota = 3.45
estudiante = "Pablo"

print("La nota de {} es {:.2f}".format(estudiante, nota))
```
